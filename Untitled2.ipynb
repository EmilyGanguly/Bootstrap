```python
import pandas as pd
import os
os.listdir('/content')
folder_path = '/content/spms-month-wise'
all_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.xlsx')]
combined_df = pd.concat((pd.read_excel(f) for f in all_files),ignore_index=True)
combined_df.head()
```

```python
combined_df
```

```python
combined_df.columns = combined_df.iloc[1]
df = combined_df[2:]
```

```python
df.head()
df = df[df['Material']!='Material']
```

```python
df['Material'].unique()
```

```python
df = df.dropna()
```

```python
list(df)
```

```python
df = df.loc[:,df.columns.notna()]
```

```python
df = df.drop(columns=['Sale QTY','Sales Organization'],errors='ignore')
```

```python
df = df.groupby(['Material','Cal. year / month','Order doc. type'],as_index=False).sum()
```

```python
df['Cal. year / month'] = pd.to_datetime(df['Cal. year / month'],format='%m.%Y')
df = df.sort_values(by='Cal. year / month')
```

```python
df
```

```python
one_material = df[df['Material']=='91307035000S']
one_material
```

```python
one_material.drop(columns=['Order doc. type'],inplace=True)
one_material= one_material.groupby(['Cal. year / month'],as_index=False).sum()
one_material = one_material.drop(columns=['Material'])
one_material.plot(x='Cal. year / month',y='Order QTY',figsize=(12,6))
one_material.head()
```

```python
df['Cal. year / month'].unique()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
```

```python
results = seasonal_decompose(one_material['Order QTY'],model='additive',period=12)
results.plot();
```

```python
len(one_material)
```

```python
train = one_material[:int(len(one_material)*0.8)]
test = one_material[int(len(one_material)*0.8):]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
one_material.head(),one_material.tail()
```

```python
train['Cal. year / month'] = train['Cal. year / month'].astype('int64')
scaler.fit(train)
scaled_train = scaler.transform(train)
test['Cal. year / month'] = test['Cal. year / month'].astype('int64')
scaled_test = scaler.transform(test)
```

```python
train
```

```python
scaled_train
```

```python
import keras
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
import numpy as np
```

```python
n_input = 3
n_features = 1
scaled_train = scaled_train[:,0].reshape(-1,1)
generator = TimeseriesGenerator(scaled_train,scaled_train,length=n_input,batch_size=1)
```

```python
x,y = generator[1]
print(f'Given the Array: \n{x.flatten()}')
print(f'Predict this y: \n {y}')
```

```python
x.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM,Dropout
```

```python
model = Sequential()
model.add(LSTM(100,activation='relu',input_shape=(n_input,n_features)))
# model.add(Dropout(0.15))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
model.fit(generator,epochs=50)
```

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
```

```python
last_train_batch = scaled_train[-n_input:]
last_train_batch = last_train_batch.reshape((1,n_input,n_features))
model.predict(last_train_batch)
```

```python
scaled_test[0]
```

```python
test_predictions = []
first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1,n_input,n_features))
```

```python
for i in range(len(test)):
  current_pred = model.predict(current_batch)[0]
  test_predictions.append(current_pred)
  current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```
